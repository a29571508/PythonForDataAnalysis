```python
grouped1=df.groupby('路線方向')
grouped1.sum()
```

```python
grouped2=df.groupby('路線方向')['週六', '週日']
sum=grouped2.sum()
#sum['週六'][sum['週六'] >= 200000]
series6 = sum['週六']
series6[series6 >= 200000]
```

```python
series7 = sum['週日']
series7[series7 >= 200000]
```

```python
series_sum=sum.stack()
series_sum
```

```python
series_result=series_sum[series_sum >= 2000000]
result=series_result.unstack()
result
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
figure1, ax1= plt.subplots()
result.plot(ax=ax1, kind='bar')
ax1.legend(prop=font)
ax1.set_xticklabels(result.index,fontproperties=font)
ax1.set_title('流量超過200萬的路線',fontproperties=font)
```
