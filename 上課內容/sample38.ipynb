```python
import pandas as pd
import requests

url='https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/json?page=0&size=1000'

response = requests.get(url)
if response.status_code == 200:
    print("資料載入成功")
    #print(response.text)
    allPythonData = response.json()
allDataFrame = pd.DataFrame(allPythonData,columns=['sna', 'tot', 'sbi', 'sarea', 'ar', 'bemp'])
allDataFrame
```
