```python
import json
import pandas as pd

with open('新北市公共自行車租賃系統.json', 'r', encoding='utf-8') as json_file:
    data = json.load(json_file)
    if data['success']:
        records = data['result']['records']
        #print(records)
    else:
        print('讀取失敗')

#columns是要選出想要的欄位
dataFrame = pd.DataFrame(records,columns=['sno', 'sarea', 'sna', 'ar', 'tot', 'sbi', 'bemp', 'lat', 'lng'])

#columns改變欄位的名稱
dataFrame.columns = ['編號', '區域', '站名', '地址', '全部數量', '可借', '可還', '緯度', '經度']
dataFrame[dataFrame['區域'] == '新店區']
```

```python
import numpy as np
all=dataFrame['全部數量'].astype(np.int64)
all.sum()
```

```python
sbi=dataFrame['可借'].astype(np.int64)
sbi.sum()
```

```python
bemp=dataFrame['可還'].astype(np.int64)
bemp.sum()
```
